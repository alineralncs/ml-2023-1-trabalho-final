```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.cluster import KMeans

X = dataset[['playlist_genre', 'playlist_subgenre', 'track_name']]
print(f'X\n {X}')

label_encoder_genre = LabelEncoder()
label_encoder_subgenre = LabelEncoder()

X_encoded = X.copy()
X_encoded['playlist_genre'] = label_encoder_genre.fit_transform(X_encoded['playlist_genre'])
X_encoded['playlist_subgenre'] = label_encoder_subgenre.fit_transform(X_encoded['playlist_subgenre'])

# Remover a coluna 'track_name'
X_encoded_no_track = X_encoded[['playlist_genre', 'playlist_subgenre']]

encoder = OneHotEncoder()
X_encoded_no_track = encoder.fit_transform(X_encoded_no_track)

num_clusters = 30
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X_encoded_no_track)

genre = 'pop'
subgenre = 'pop edm'

new_playlist_genre = label_encoder_genre.transform([genre])
new_playlist_subgenre = label_encoder_subgenre.transform([subgenre])

new_playlist_genre_encoded = encoder.transform([[new_playlist_genre[0], new_playlist_subgenre[0]]]).toarray()

# cluster_label = kmeans.predict(new_playlist_genre_encoded)
cluster_label = kmeans.predict(new_playlist_genre_encoded)

# Filtrar os exemplos no conjunto de dados original que pertencem ao cluster identificado
cluster_samples = X[kmeans.labels_ == cluster_label[0]]

num_recommendations = 10
recommendations = cluster_samples.sample(n=num_recommendations)[['track_name', 'playlist_genre', 'playlist_subgenre']]

print('Lista de recomendações:')
for index, row in recommendations.iterrows():
    print('Música:', row['track_name'])
    print('Gênero:', row['playlist_genre'])
    print('Subgênero:', row['playlist_subgenre'])
    print('---')
# Informações sobre os centroides
print('Centroides:')
print(kmeans.cluster_centers_)
print(kmeans.labels_)

# Informações sobre as iterações
print('Número de iterações:', kmeans.n_iter_)
```

```python
# Adicione essa linha antes de plotar o gráfico
X['cluster_label'] = kmeans.labels_

# Decodificar os valores numéricos de 'playlist_genre' para as categorias originais
X['playlist_genre_decoded'] = label_encoder_genre.fit_transform(X['playlist_genre'])

plt.figure(figsize=(10, 6))
cluster_counts = X.groupby(['playlist_genre_decoded', 'cluster_label']).size().unstack(fill_value=0)
cluster_counts.plot(kind='bar', stacked=True, colormap='viridis')
plt.xlabel('Playlist Genre')
plt.ylabel('Count')
plt.title('Songs Clustered by Genre')
plt.xticks(rotation=45)
plt.legend(title='Cluster Label')
plt.show()
```

```python
import matplotlib.cm as cm


X['cluster_label'] = kmeans.labels_

X['music'] = label_encoder_genre.fit_transform(X['track_name'])


custom_cmap = cm.get_cmap('tab20')

plt.figure(figsize=(10, 6))
plt.scatter(x=X['playlist_genre'], y=X['music'], c=X['cluster_label'], cmap=custom_cmap)
plt.xlabel('Subgênero da Playlist')
plt.ylabel('Músicas')
plt.title('Músicas Agrupadas por genero')
#plt.xticks(np.arange(len(label_encoder_subgenre.classes_)), label_encoder_subgenre.classes_, rotation=90)
plt.colorbar(label='Rótulo do Cluster')
plt.show()
```
